## Feature Processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline 
```

```python
X.head()
```

```python
numeric = ['usd_goal_real','daysavailable']
cat = ['main_category','currency']#,'year']

numpipe = Pipeline([
        ('nummiss', SimpleImputer(missing_values=np.nan, strategy="mean")),
        ('std_scaler', StandardScaler()
        )])

catpipe = Pipeline([
        ('catmiss', SimpleImputer(missing_values=np.nan, strategy="most_frequent")),
        ('hot', OneHotEncoder(drop="first")
        )])

transform = ColumnTransformer(transformers=[
        ('num', numpipe, numeric),
        ('cat', catpipe, cat)],
        remainder='drop')
```

```python
X_small = X[0:50]
X_small.head()
```

```python
finaldata = transform.fit_transform(X)
#finaldata = pd.DataFrame(finaldata.toarray())
finaldata = pd.DataFrame(finaldata.toarray(),
                        columns=(numeric +
                                list(transform.named_transformers_['cat']['hot'].get_feature_names(cat))))

finaldata
```

## Hyperparameter Tuning

### Random Forest

```python
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

```python
# split the data
X_train, X_test, y_train, y_test = train_test_split(finaldata,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=123)
```

```python
max_depth = ['None',1,5,15,20]
n_estimators = [10,50,100,200]
df = pd.DataFrame([[0 for x in range(len(n_estimators))] for y in range(len(max_depth))])
df.columns=n_estimators
df.index=max_depth
df
```

```python
for i in max_depth:
    for j in n_estimators:
        if i=='None':
            RF = RandomForestClassifier(n_estimators=j, max_depth=None, random_state=123).fit(X_train[0:200], y_train[0:200])
        else:
            RF = RandomForestClassifier(n_estimators=j, max_depth=i, random_state=123).fit(X_train[0:200], y_train[0:200])
        df.loc[i,j]= 1 - RF.score(X_train, y_train)
df
```

```python
for i in n_estimators:
    plt.plot(max_depth, df[i],label=n_estimators[df.columns.get_loc(i)])
    plt.legend(loc="upper right", markerscale= 0.1)
    plt.xlabel("Max-depth", fontsize="10")
    plt.ylabel("Train error", fontsize="10")
    plt.title("Train error under different max depths",fontsize="15") 
```

### KNN model

```python
train_error=[]
n_neighbors=range(1,15)
for i in n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=i).fit(X_train[0:200], y_train[0:200])
    train_error.append(1 - knn.score(X_train[0:200], y_train[0:200]))
```

```python
plt.plot(n_neighbors, train_error)
plt.xlabel("Number of neighbors", fontsize="10")
plt.ylabel("Train error", fontsize="10")
plt.title("Train error under different neighbors",fontsize="15") 
plt.show()
```

## Model Selection

```python
#fit the model
pipe = Pipeline([('classifier', LogisticRegression(solver='lbfgs'))])
search_space = [{'classifier': [LogisticRegression(solver='lbfgs')],
                 'classifier__C': [0.01, 0.1, 1.0]},
                {'classifier': [RandomForestClassifier(random_state = 123)],
                 'classifier__n_estimators': [10, 50, 100, 150,500],
                 'classifier__max_depth': [None,10, 20]},
                {'classifier': [KNeighborsClassifier()],
                 'classifier__n_neighbors':[1,10,20]}]                
                
clf = GridSearchCV(pipe, search_space, cv=5, verbose=10,n_jobs=-1)
best_model=clf.fit(X_train,y_train)
```
